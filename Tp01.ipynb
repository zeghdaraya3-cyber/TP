```python
# -----------------------------
# TP OpenScience - Analyse CSV
# -----------------------------

import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. Définir le dossier du projet
# -----------------------------
os.chdir(r"C:\Users\zizou\Downloads\tp_openscience01-main\tp_openscience01-main")

# -----------------------------
# 2. Lire le fichier CSV
# -----------------------------
df = pd.read_csv("data_biology.csv")

# Afficher les 5 premières lignes
print("Aperçu des données :")
print(df.head())

# Afficher les colonnes
print("\nColonnes disponibles :")
print(df.columns)

# -----------------------------
# 3. Statistiques descriptives
# -----------------------------
print("\nStatistiques descriptives (moyenne et écart-type) :")
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    mean = df[col].mean()
    std = df[col].std()
    print(f"{col} -> Moyenne : {mean:.2f}, Écart-type : {std:.2f}")

# -----------------------------
# 4. Graphiques
# -----------------------------
sns.set(style="whitegrid")

# 4.1 Cell count par sample et group
plt.figure(figsize=(12,6))
sns.barplot(x='sample', y='cell_count', hue='group', data=df)
plt.xticks(rotation=45)
plt.title("Cell Count par Sample et Group")
plt.xlabel("Sample")
plt.ylabel("Cell Count")
plt.tight_layout()
plt.show()

# 4.2 Growth rate selon temperature et group
plt.figure(figsize=(8,5))
sns.lineplot(x='temperature', y='growth_rate', hue='group', marker='o', data=df)
plt.title("Growth Rate selon la Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Growth Rate")
plt.show()

# 4.3 Protein level par sample et group
plt.figure(figsize=(12,6))
sns.barplot(x='sample', y='protein_level', hue='group', data=df)
plt.xticks(rotation=45)
plt.title("Protein Level par Sample et Group")
plt.xlabel("Sample")
plt.ylabel("Protein Level")
plt.tight_layout()
plt.show()

# 4.4 pH et oxygen selon temperature (axes séparés)
fig, ax1 = plt.subplots(figsize=(10,6))

# Axe gauche pour pH
sns.lineplot(x='temperature', y='ph', hue='group', marker='o', data=df, ax=ax1)
ax1.set_ylabel("pH")
ax1.set_xlabel("Temperature (°C)")

# Axe droit pour oxygen
ax2 = ax1.twinx()
sns.lineplot(x='temperature', y='oxygen', hue='group', marker='x', data=df, ax=ax2, legend=False)
ax2.set_ylabel("Oxygen")

plt.title("pH et Oxygen selon la Temperature")
plt.show()

# -----------------------------
# 5. Graphiques automatiques pour toutes les colonnes numériques
# -----------------------------
for col in numeric_cols:
    plt.figure(figsize=(8,5))
    sns.boxplot(x='group', y=col, data=df)
    plt.title(f"Distribution de {col} par group")
    plt.show()
```
